```python
merged1=merged.merge(excel_carrera, on=['ID_Carrera'], suffixes=('_left','_right'))
```

```python
merged1
```

```python
#La mejor(es) nota final del anio
merged1[['Nombres','Nombre_Carrera','Nota_final']][merged1.Nota_final==merged1['Nota_final'].max()]
```

```python
#La cantidad de alumnos que aprobaron su anio con nota mayor o igual a 65%
merged1[merged1.Nota_final>=65].count()['ID_Estudiante'] 

```

```python
#Los alumnos de la carrera de salud que pasaron su anio escolar
merged1[merged1.Nombre_Carrera=='Salud']
```

```python
#Media de las notas finales del anio
merged1['Nota_final'].median()
```
